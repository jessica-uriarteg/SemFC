```python
# Cargamos la base de datos, ya editada en emacs (si es necesario) y lista para trabajarse
df=pd.read_csv('muestra.txt',names=['Ruido'], sep=',')
# Le damos la estructura de datos
df=pd.DataFrame(df)
# Vemos la base de datos empezando de arriba (head)
df.head()
```

```python
# Creamos una lista mylist con los datos del archivo muestra.txt
mylist=[]
with open("muestra.txt") as f:
    for line in f:
        mylist.append(float(line.strip()))
        
print(mylist)
```

```python
# Graficamos los datos tales como los bajamos
plt.hist(mylist,50,normed=1,facecolor='c',alpha=0.75)
plt.title("Grafica con datos sucios")
plt.xlabel("Ruido atmosférico")
plt.ylabel("Entradas")
plt.show()
```

```python
# Creamos una nueva base de datos con los datos limpios, borramos los datos duplicados dejando el primer dato intacto 
df1=df.drop_duplicates(subset=None, keep='first', inplace=False)
```

```python
# Graficamos los datos limpios
df1.hist(normed=1,facecolor='c',bins=50)
plt.title("Gráfica con datos limpios")
plt.xlabel("Ruido atmosférico")
plt.ylabel("Entradas")
plt.show()
```

```python
# .describe() me da la descripción estadística de los datos limpios. 
# Tengo 1502 datos en total, media= 10.007, desviación estándar=0.293157
# 25%, 50% y 75% nos da los cuartiles
df1.describe()
```

```python
# Realizamos el ajuste y almacena el resultado en las variables mu y std
mu,std=norm.fit(df1)
# Imprimimos los valores
print(std)
print(mu)
```

```python
# Extraemos los elementos del histograma, definimos el PDF ya con los parámetros de ajuste e imprimos el plt
n,bins,patches=plt.hist(df1["Ruido"],facecolor='c',bins=50,normed=1)
y=norm.pdf(bins,mu,std)
plt.plot(bins,y,'r--',linewidth=2)
plt.title("Gráfica con datos limpios normalizada con ajuste de Gaussiana")
plt.xlabel("Ruido atmosférico")
plt.ylabel("Entradas")
plt.show()
```
